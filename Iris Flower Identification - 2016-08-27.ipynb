```python
# IRIS Dataset
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
n_samples, n_features = iris.data.shape
print('Number of samples : ', n_samples)
print('Number of features : ', n_features)
print(iris.data[0])

print('target :', iris.target.shape)
print('iris target  : ', iris.target)
print('iris names :', iris.target_names)

print('samples of target distribution : ',np.bincount(iris.target))

arr = np.random.randint(0,10,10000)
print('samples of random distribution : ', np.bincount(arr))


```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x_index = 3
colors = ['blue', 'red', 'green']

labels = [1,2,3]
lc = zip(labels, colors) # Warning, zip (as range), is a generator that can be used only once.
print(list(lc))

lc = zip(labels, colors)
for label, color in lc :
    print('label : {}, color : {}'.format(label,color))

# Now display iris with the help of this function
for label, color in zip(range(len(iris.target_names)), colors):
    print('label : {}, color : {}'.format(label,color))
    plt.hist(iris.data[iris.target==label, x_index],
             label = iris.target_names[label],
             color=color)
    plt.xlabel(iris.feature_names[x_index])
    plt.legend(loc='upper right')
    plt.show
    
```
