```python
# IRIS Dataset
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
n_samples, n_features = iris.data.shape
print('Number of samples : ', n_samples)
print('Number of features : ', n_features)
print(iris.data[0])

print('target :', iris.target.shape)
print('iris target  : ', iris.target)
print('iris names :', iris.target_names)

print('samples of target distribution : ',np.bincount(iris.target))

arr = np.random.randint(0,10,10000)
print('samples of random distribution : ', np.bincount(arr))


```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x_index = 3
colors = ['blue', 'red', 'green']

labels = [1,2,3]
lc = zip(labels, colors) # Warning, zip (as range), is a generator that can be used only once.
print(list(lc))

lc = zip(labels, colors)
for label, color in lc :
    print('label : {}, color : {}'.format(label,color))

# Now display iris with the help of this function
for label, color in zip(range(len(iris.target_names)), colors):
    print('label : {}, color : {}'.format(label,color))
    plt.hist(iris.data[iris.target==label, x_index],
             label = iris.target_names[label],
             color=color)
    plt.xlabel(iris.feature_names[x_index])
    plt.legend(loc='upper right')
    plt.show
    
```

```python
# using sklearn
from sklearn import datasets
from sklearn.datasets import get_data_home

get_data_home()

from sklearn.datasets import load_digits
digits = load_digits()

n_samples, n_features = digits.data.shape
print('Number of samples : ', n_samples)
print('Number of features : ', n_features)
print(digits.data[0])
print(digits.images[0])
print(digits.target)

print('image:', digits.images[0].reshape(64))
if np.array_equal( digits.data, digits.images.reshape(1797,64)):
    print('These are the same dataset')
else:
    print('datasets differ')
    
fig = plt.figure(figsize=(10,10))

for i in range(64):
    ax = fig.add_subplot(8,8,i+1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap = plt.cm.binary, interpolation = 'nearest')
    
    #label the image with the target values
    ax.text(6,7,digits.target_names[digits.target[i]])

fig.show()
    
```

```python
#S-Curve

from sklearn.datasets import make_s_curve
data, colors = make_s_curve(n_samples=1000)
print(data.shape)
print(colors.shape)

print(data)
#print(colors)

fig = plt.figure(figsize=(10,10))
plt.scatter(data[:,0], data[:,1])

from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2], c = colors)
ax.view_init(10,-60)

```

```python
# Faces

from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()

print(faces.data)
print(faces.data.shape)

print(faces.images.shape)

fig = plt.figure(figsize=(10,10))

# display just a subset of the 400 images for time and performance... 
for i in range(16):
    ax = fig.add_subplot(4,4,i+1, xticks=[], yticks=[])
    ax.imshow(faces.images[i], cmap = plt.cm.bone, interpolation = 'nearest')
    
    #label the image with the target values
    ax.text(6,7,digits.target_names[digits.target[i]])

fig.show()
```
