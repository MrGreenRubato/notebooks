# MNIST Loading

Using basic functionality to load mnist training set.

Tutorial used : http://deeplearning.net/tutorial/deeplearning.pdf



```python
%matplotlib inline
import cPickle, gzip, numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython import display

```

```python
# Load the dataset
f = gzip.open('mnist/mnist.pkl.gz', 'rb')
train_set, valid_set, test_set = cPickle.load(f)
f.close()
```

## Display images as ASCII Code

```python
for im in range (0,10):
    image = ''
    img = train_set[0][im]
    for i in range(len(img)):
        if i%28==0 : 
            image += '\n'
        if img[i] > .500:
            image += '@'
        elif img[i] > .200:
            image += '+'
        elif img[i] > .01:
            image += '.'
        else:
            image += ' '
    print(image)
```

## display image with pyplot or any image library in grayscale


```python
imageCount = 1000
imageSize = 28
width = np.ceil(np.sqrt(imageCount)).astype(int)
height = np.ceil(np.sqrt(imageCount)).astype(int)
print(width)
png = np.zeros((width*28,height*28,3))
for im in range (0,imageCount):
    img = train_set[0][im]
    for i in range(len(img)):
        line = (im/height)*28;
        col = (im%width)*28;
        png[(i/28)+line][(i%28)+col][0] = img[i]
        png[(i/28)+line][(i%28)+col][1] = img[i]
        png[(i/28)+line][(i%28)+col][2] = img[i]    
#imgplot = plt.imshow(png)
plt.figure(figsize=(imageCount/20,imageCount/20))
plt.imshow(png)
```
