```python
# Loading JSON to python code and display it

import json, datetime
from pprint import pprint

with open('temp.json') as data_file:    
    data = json.load(data_file)
    
#pprint(data)

data[1:10]

import numpy as np
temp = np.array([], 'float')
calc = np.array([], 'float')
amb = np.array([], 'float')
date = np.array([], dtype='datetime64[ms]')



for d in data:
    temp = np.insert(temp,0,d['temperature'])
    calc = np.insert(calc,0,d['calculated'])
    amb = np.insert(amb,0,d['ambient'])
    #d = datetime.datetime.strptime(d['date'], '%Y-%m-%d %H:%M:%S')
    dt = np.datetime64(d['date'])
    date = np.insert(date, 0, dt)
    
import matplotlib.pyplot as plt
%matplotlib inline

last_hours = 24*7
#sample is 5 minutes
extend = last_hours * 60 // 5
plt.figure(figsize=(12,6))
plt.plot(date[-extend::12], temp[-extend::12], 'b', date[-extend::12],amb[-extend::12], 'r',  date[-extend::12],calc[-extend::12], 'g', )
plt.tight_layout()
```
