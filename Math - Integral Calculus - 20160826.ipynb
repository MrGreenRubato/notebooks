```python
# Integral calculus : finding area under a function
import math
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,np.pi*2,100)

def f(x):
    return np.cos(x)
    #return x**2 + 2
    #return x**3 + x**2 + x + 1

%matplotlib inline
plt.figure(figsize=(5,5))
plt.plot(x, f(x))


```

```python
# surface [0,5]
dx = 0.001
startx = 0
endx = np.pi*2
area = 0
y = np.array([])

breadth = np.arange(startx, endx, dx)

for i in breadth:
    y = np.append(y,f(i+dx))
    area = area + f(i+dx)*dx


plt.bar(breadth,y,dx, alpha=0.2)
plt.plot(x, f(x))
plt.show()

print(area)
```
