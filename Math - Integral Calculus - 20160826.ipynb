```python
# Integral calculus : finding area under a function
import math
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,np.pi*2,100)

def f(x):
    #return x**2 + 2
    #return x**3 + x**2 + x + 1
    #return np.cos(x)
    return 3*x**(1.3) - 3*x*np.cos(x*2) + 1

%matplotlib inline
plt.figure(figsize=(6,5))
plt.plot(x, f(x))


```

```python
# surface [0,5]
MIDDLE_APPROX = True # True for far better approximation
dx = 0.3
startx = 1
#endx = np.pi*2
endx = 4
area = 0

y = np.array([])

breadth = np.arange(startx, endx, dx)

div = 1
if MIDDLE_APPROX:
    div = 2

for i in breadth:
    y = np.append(y,f(i+dx/div))
    area = area + f(i+dx/div)*dx

plt.figure(figsize=(6,5))
plt.bar(breadth,y,dx, alpha=0.2)
plt.plot(x, f(x))
plt.show()

print(area)
```
